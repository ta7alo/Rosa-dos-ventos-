#Importando bibliotecas

-----

```python
%matplotlib inline
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import matplotlib.cm as cm
from math import pi
from windrose import WindroseAxes
```

#Criando data-frame com velocidade e direção do vento

-----



```python
N = 1000
ws = np.random.random(N)*6
wd = np.random.random(N)*360
df = pd.DataFrame({"speed": ws, "direction": wd})
df
```

# Rosa dos Ventos como um Histograma 
----

```python
#Eixo da rosa dos ventos
ax = WindroseAxes.from_ax()
#Plotando as informações do vento
ax.bar(df.direction, df.speed, normed=True, opening=0.8, edgecolor='white')
#Legenda
ax.set_legend(title='Velocidade (m/s)', loc=(0.98, 0), shadow=True)
#Título do gráfico
plt.title("Rosa dos ventos como um histograma", y=1.08, fontsize='20')
```

#Rosa dos ventos com preenchimento

```python
#Eixo da rosa dos ventos
ax = WindroseAxes.from_ax()
#Plotando as informações do vento
ax.contourf(df.direction, df.speed, bins=np.arange(0, 8, 1), cmap=cm.hot)
#Legenda
ax.set_legend(title='Velocidade (m/s)', loc=(0.99, 0), shadow=True)
#Título do gráfico
plt.title("Rosa dos ventos com preenchimento", y=1.08, fontsize='20')
```

#Rosa dos ventos com linhas de contorno

```python
#Eixo da rosa dos ventos
ax = WindroseAxes.from_ax()
#Plotando as informações do vento
ax.contour(df.direction, df.speed, bins=np.arange(0, 8, 1), cmap=cm.hot, lw=3)
#Legenda
ax.set_legend(title='Velocidade (m/s)', loc=(0.99, 0), shadow=  True)
#Título do gráfico
plt.title("Rosa dos ventos com linhas de contorno", y=1.08, fontsize='20')
```
